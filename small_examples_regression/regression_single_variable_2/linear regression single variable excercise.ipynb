```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df=pd.read_csv("Desktop\\canada_per_capita_income.csv")
df
```

```python
p1=df['year']
p1
```

```python
p2=df['per capita income (US$)']
p2
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('pci(US$)')
plt.scatter(p1,p2,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],p2)
```

```python
reg.predict(2020)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('pci(US$)')
plt.scatter(p1,p2,color='red',marker='+')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
