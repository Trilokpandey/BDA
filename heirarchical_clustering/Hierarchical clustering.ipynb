## Load Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import normalize
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

## Loading of dataset through pandas

```python
df=pd.read_csv("F:\heirarchical_clustering\Wholesale customers data.csv")
df.head()
```

## scaling of dataframe

```python
scaled_df=normalize(df)
scaled_df=pd.DataFrame(scaled_df,columns=df.columns)
scaled_df.head()
```

## plotting the cluster on dendrograms

```python
plt.figure(figsize=(10,7))
plt.title("dendrograms")
dend=shc.dendrogram(shc.linkage(scaled_df,method='ward'))
```

x-asis contain the data sample and y-axis represent the distance between these samples

```python
plt.figure(figsize=(10,7))
plt.title("Dendrograms")
dend=shc.dendrogram(shc.linkage(scaled_df,method='ward'))
plt.axhline(y=6,color='r',linestyle='--')
```

we drew the line on dendrogram to get the number of clusters.

```python
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit_predict(scaled_df)
```

The value o represent the points belong to cluster 1 and the value 1 represent the points belong to cluster 2.

## plotting of cluster

```python
plt.figure(figsize=(10,7))
plt.scatter(scaled_df['Milk'],scaled_df['Grocery'],c=cluster.labels_)
```
